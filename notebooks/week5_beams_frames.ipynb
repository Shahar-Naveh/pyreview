## 5.1 Simply Supported Beam
Pin at A (x=0), Roller at B (x=L)

```python
# Example: Beam with point load and distributed load
L = 8.0  # [m]
loads = [
    {'type': 'point', 'P': 20, 'x': 3.0},           # 20 kN at x=3m
    {'type': 'distributed', 'w': 5, 'x_start': 4.0, 'x_end': 8.0},  # 5 kN/m from x=4 to x=8
]

x, V, M, reactions = shear_moment_simply_supported(L, loads)

print(f"Reactions: Ay = {reactions['Ay']:.2f} kN, By = {reactions['By']:.2f} kN")

extremes = find_max_shear_moment(x, V, M)
print(f"V_max = {extremes['V_max']:.2f} kN at x = {extremes['V_max_x']:.2f} m")
print(f"M_max = {extremes['M_max']:.2f} kN·m at x = {extremes['M_max_x']:.2f} m")

plot_sfd_bmd(x, V, M, title=f"Simply Supported Beam (L = {L} m)")
```

## 5.2 Cantilever Beam

```python
# Practice: Cantilever with tip load
L = 4.0  # [m]
loads = [
    {'type': 'point', 'P': 15, 'x': 4.0},  # 15 kN at free end
]

x, V, M, reactions = shear_moment_cantilever(L, loads)
print(f"Fixed-end reactions: Ay = {reactions['Ay']:.2f} kN, MA = {reactions['MA']:.2f} kN·m")
plot_sfd_bmd(x, V, M, title="Cantilever Beam — Tip Load")
```
