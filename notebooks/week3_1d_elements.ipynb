## 3.1 Basic Bar Analysis

## 3.3 Thermal Effects
Temperature changes cause strain and stress (if restrained).

```python
# Example: Restrained steel bar heated by 40°C
E_steel = 200e9      # [Pa]
alpha_steel = 12e-6   # [1/°C]
delta_T = 40          # [°C]

eps_T = thermal_strain(alpha_steel, delta_T)
sigma_T = thermal_stress_restrained(E_steel, alpha_steel, delta_T)

print(f"Thermal strain: {eps_T:.6e}")
print(f"Thermal stress: {sigma_T/1e6:.2f} MPa (compressive)")
```
