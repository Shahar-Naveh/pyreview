## 1.4 Numpy Arrays
Vectorized operations — the foundation of engineering computing.

```python
# Example: Vectorized stress calculation (no loops needed)
forces = np.array([10e3, 25e3, 50e3, 75e3, 100e3])  # [N]
A = 0.01  # [m²]
stresses = forces / A  # [Pa] — operates on entire array at once
print("Stresses [MPa]:", stresses / 1e6)
```
