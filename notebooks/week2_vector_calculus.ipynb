## 2.1 Force Vectors
Represent forces as numpy arrays.

```python
# Example: Two forces acting on a point
F1 = np.array([3.0, 4.0])   # [kN]
F2 = np.array([-1.0, 2.0])  # [kN]

# Resultant
R = F1 + F2
R_mag = np.linalg.norm(R)
print(f"Resultant: R = {R} kN")
print(f"Magnitude: |R| = {R_mag:.2f} kN")
print(f"Direction: θ = {np.degrees(np.arctan2(R[1], R[0])):.1f}°")
```

## 2.2 Moment About a Point (2D)
M = r × F (scalar in 2D: M = rx·Fy − ry·Fx)

```python
# Practice: Compute moment of a force about point A
F = np.array([0, -10])  # 10 kN downward
position = np.array([3, 0])  # applied at x = 3 m
point_A = np.array([0, 0])  # moment about origin

M = moment_about_point(F.reshape(1, 2), position.reshape(1, 2), point_A)
print(f"Moment about A = {M:.2f} kN·m")
```

## 2.3 3D Force Systems
Cross products and moment vectors.

```python
# Practice: 3D moment calculation
F_3d = np.array([[0, 0, -10]])  # 10 kN downward [kN]
pos_3d = np.array([[3, 2, 0]])  # application point [m]
origin = np.array([0, 0, 0])

M_3d = moment_about_point_3d(F_3d, pos_3d, origin)
print(f"Moment vector: M = {M_3d} kN·m")
```
