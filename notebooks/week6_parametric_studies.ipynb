## 6.1 Single-Parameter Study

```python
# Example: Effect of cross-sectional area on stress
P = 100e3  # axial force [N]
areas = np.linspace(100e-6, 2000e-6, 50)  # 100 to 2000 mm² in m²

stresses = axial_stress(P, areas) / 1e6  # [MPa]

plot_parametric(areas * 1e6, stresses,
                param_name='Cross-sectional Area A [mm²]',
                result_name='Stress σ [MPa]',
                title='Stress vs. Cross-sectional Area (P = 100 kN)')
```

## 6.2 Multi-Parameter Study with Pandas

```python
# Example: Compare elongation for different materials and lengths
materials = {
    'Steel':    {'E': 200e9, 'color': 'blue'},
    'Aluminum': {'E': 70e9,  'color': 'orange'},
    'Concrete': {'E': 30e9,  'color': 'gray'},
}

P = 50e3      # [N]
A = 500e-6    # [m²]
lengths = np.linspace(1, 10, 20)  # [m]

fig, ax = plt.subplots(figsize=(10, 5))
results = []

for name, props in materials.items():
    deltas = elongation(P, lengths, A, props['E']) * 1e3  # [mm]
    ax.plot(lengths, deltas, 'o-', color=props['color'], label=name, linewidth=2)
    for L_val, d_val in zip(lengths, deltas):
        results.append({'Material': name, 'L [m]': L_val, 'δ [mm]': d_val})

ax.set_xlabel('Length L [m]')
ax.set_ylabel('Elongation δ [mm]')
ax.set_title('Elongation vs. Length for Different Materials')
ax.legend()
ax.grid(True)
plt.tight_layout()
plt.show()

# Summary table
df = pd.DataFrame(results)
print(df.groupby('Material')['δ [mm]'].describe().round(4))
```
