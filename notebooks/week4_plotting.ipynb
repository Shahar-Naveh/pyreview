## 4.1 Basic Engineering Plot
Every plot needs: title, axis labels with units, grid.

```python
# Example: Stress-strain curve for steel
E = 200e3  # [MPa]
sigma_y = 250  # yield stress [MPa]
eps = np.linspace(0, 0.003, 100)
sigma = np.minimum(E * eps, sigma_y)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(eps * 100, sigma, 'b-', linewidth=2)
ax.set_xlabel('Strain ε [%]')
ax.set_ylabel('Stress σ [MPa]')
ax.set_title('Stress-Strain Curve — Structural Steel')
ax.grid(True)
ax.axhline(y=sigma_y, color='r', linestyle='--', label=f'σ_y = {sigma_y} MPa')
ax.legend()
plt.tight_layout()
plt.show()
```

## 4.2 Using the Library Plot Functions

```python
# Example: Plot a simple shear force diagram using the library
x = np.linspace(0, 6, 500)
V = 10 - 5 * x  # linear shear [kN]
M = 10 * x - 2.5 * x**2  # parabolic moment [kN·m]

plot_sfd_bmd(x, V, M, title="Example Beam — Uniform Load")
```

## 4.3 Parametric Plots

```python
# Example: How does beam deflection change with length?
lengths = np.linspace(2, 10, 20)  # [m]
w = 5  # uniform load [kN/m]
E = 200e6  # [kPa]
I = 8.33e-4  # moment of inertia [m⁴]

# Max deflection for simply supported beam: δ = 5wL⁴/(384EI)
deflections = 5 * w * lengths**4 / (384 * E * I) * 1000  # [mm]

plot_parametric(lengths, deflections,
                param_name='Beam Length L [m]',
                result_name='Max Deflection δ [mm]',
                title='Deflection vs. Beam Length')
```
